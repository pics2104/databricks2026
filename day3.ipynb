```python
lineitemdf=spark.read.csv(path+"/lineitem.csv",header=True,inferSchema=True,sep="\t")
orderdf=spark.read.csv(path+"/orders.csv",header=True,inferSchema=True,sep="\t")
display(lineitemdf)
display(orderdf)
```

```python
print(customerdf.count())
print(lineitemdf.count())
print(orderdf.count())
```

```python
from pyspark.sql.functions import *
orders_cus_df=orderdf.join(broadcast(customerdf), orderdf.O_CUSTKEY == customerdf.C_CUSTKEY, "inner").explain()
```
